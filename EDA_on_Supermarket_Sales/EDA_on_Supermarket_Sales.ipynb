### a) Sales by Branch

```python
branch_sales = data.groupby('Branch')['Total'].sum()  # Aggregates total sales for each branch.
print(branch_sales)
```

### b) Sales by Payment

```python
Payment_sales = data.groupby('Payment')['Total'].sum()  # Aggregates total sales for each payment.
print(Payment_sales)
```

### c) Sales by Rating

```python
Rating_sales = data.groupby('Rating')['Total'].sum()  # Aggregates total sales for each Rating.
print(Rating_sales)
```

### Sales by Product Line

```python
product_sales = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print(product_sales)
```

`groupby('Product line'):` Groups data by product categories.

`['Total'].sum():` Sums up sales for each product line.

`sort_values(ascending=False):` Arranges the product lines from highest to lowest sales.

### Gender-Based Spending Behavior

```python
gender_spending = data.groupby('Gender')['Total'].mean()  # Average spending per gender.
print(gender_spending)
```

`groupby('Gender'): `Groups customers by gender (Male/Female).

`['Total'].mean(): `Calculates the average spending for each gender.



### Time-Based Analysis

```python
data['Month'] = data['Date'].dt.month  # Extracts the month from the 'Date' column.

monthly_sales = data.groupby('Month')['Total'].sum()  # Calculates total sales per month.
print(monthly_sales)
```

`data['Date'].dt.month:` Extracts the month (1-12) from the Date column and creates a new Month column.

`groupby('Month')['Total'].sum():` Groups data by month and calculates total sales.

# 4. Data Visualization
### a) Pie chart: Sales Distribution by Branch

```python
branch_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Branch', figsize=(6, 8))
plt.ylabel('')  # Removes the y-axis label for cleaner presentation.
plt.show()
```

`plot(kind='pie'):` Creates a pie chart for branch sales.

`autopct='%1.1f%%':` Displays percentages on the pie chart.

`figsize=(8, 8):` Sets the figure size to 8x8 inches.

### b) Bar chart: Payment methods vs. total sales

```python
sns.barplot(x=product_sales.values, y=product_sales.index,hue=product_sales, palette='viridis',saturation=.6)
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.show()
```

`sns.barplot():` Creates a horizontal bar chart.

`x=product_sales.values: `Plots sales values on the x-axis.

`y=product_sales.index:` Uses product lines on the y-axis.

`palette='viridis':` Chooses a color palette for the bars.


### c) Heatmap: Correlation between numerical features

```python
# Calculate the correlation matrix for numerical features
correlation_matrix = data.corr(numeric_only=True)  # Use numeric_only=True for pandas v2.0+

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)

# Add a title
plt.title("Correlation Heatmap Between Numerical Features")

# Display the heatmap
plt.show()
```

`data.corr(numeric_only=True):`Calculates the pairwise correlation of all numerical columns in the dataset.

`sns.heatmap:`Visualizes the correlation matrix as a heatmap.

`annot=` True adds the correlation values inside the heatmap.

`cmap= `"coolwarm" provides a color gradient for the correlation values.

`plt.figure(figsize=(10, 6)): `Adjusts the size of the heatmap.

### d) Boxplot: Sales distribution by customer gender

```python
genders = data.groupby('Gender')['Total'].sum()  
genders
```

```python
sns.boxplot(x='Gender', y='Total',data= data,hue='Gender',palette=['m','c'],saturation=.5)

plt.show()
```

`palette=['m', 'c']: ` Specifies the colors for the genders ('r' for red, 'g' for green).

`plt.show():` Displays the plot.
